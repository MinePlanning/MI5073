## Función para importar el Modelo de Bloques.

Existen varias formas de leer un archivo separado por comas y almacenarlo en memoria. Librerías como $Numpy$ y $Pandas$ tienen implementaciones eficientes para estos fines. El ejemplo a continuación lee el archivo de texto y lo almacena en un arreglo de arreglos (cada elemento del arreglo es un arreglo de 1D).

```python
def importGTBlockModel(filename):
    blockModel=[];
    file = open(filename);
    for ln in file:              
        array=[];
        line=ln.split(',');          
        grade=float(line[0]);
        ton=float(line[1]);
        array.append(grade);
        array.append(ton);
        blockModel.append(array);
    file.close();
    return blockModel;    
```

## Ejemplo del uso de la función $importGTBlockModel$

```python
gtBM=importGTBlockModel("MI5073_GT_1.csv");
```
